# MDP Value and Policy Iteration Analysis

1. Value Iteration: $ V_{k+1}{(s)} = \max_{a}[R(a,s) + \gamma \sum_{s'}{P(s'|a,s)V^*(s')}] $
    - Standard Value Iteration
    - Gauss-Seidel Value Iteration
    - Prioritised Sweeping Value Iteration

2. Policy Iteration: $ V_{k+1}{(s)} = R(\pi(s),s) + \gamma \sum_{s'}{P(s'|\pi(s),s)V^*(s')} $

```python
import gym

import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from agents import PolicyIterationAgent, ValueIterationAgent, ValueIterationGSAgent, ValueIterationPSAgent
```

## Frozen Lake

- Agent controls movement over a grid
- Some tiles walkable, some lead agent to fall into water; agent has a start and goal tile
- Movement of agent with uncertainty (due to slippery ice)
- Rewards: +1 if agent finds correct path, 0 otherwise

```python
env_name = "FrozenLake8x8-v1"

env = gym.make(env_name)

agent = ValueIterationAgent(env, gamma=0.9)

print(f"Number of iterations until convergence using {agent}: {len(agent.updates)}")

sns.lineplot(x=range(len(agent.updates)), y=agent.updates)

plt.title(f"{env_name} Convergence Graph")
plt.xlabel("Iteration")
plt.ylabel("||V-V*||2")
plt.show()
```

```python
env_name = "FrozenLake8x8-v1"

env = gym.make(env_name)

agent = ValueIterationGSAgent(env, gamma=0.9)

print(f"Number of iterations until convergence using {agent}: {len(agent.updates)}")

sns.lineplot(x=range(len(agent.updates)), y=agent.updates)

plt.title(f"{env_name} Convergence Graph")
plt.xlabel("Iteration")
plt.ylabel("||V-V*||2")
plt.show()
```

```python
env_name = "FrozenLake8x8-v1"

env = gym.make(env_name)

agent = PolicyIterationAgent(env, gamma=0.9)

print(f"Number of iterations until convergence using {agent}: {len(agent.updates)}")

sns.lineplot(x=range(len(agent.updates)), y=agent.updates)

plt.title(f"{env_name} Convergence Graph")
plt.xlabel("Iteration")
plt.ylabel("||V-V*||2")
plt.show()
```

```python
env_name = "FrozenLake8x8-v1"

env = gym.make(env_name)

agent_updates = []

for _ in range(10):
    agent = PolicyIterationAgent(env, gamma=0.9)
    agent_updates.append(len(agent.updates))
    
print(f"Average number of iterations until convergence using {agent}: {np.mean(agent_updates)}")
```
